```python
import pandas as pd
```

```python
day_1 = pd.read_csv("dataset/02-14-2018.csv")
day_2 = pd.read_csv("dataset/02-15-2018.csv")
day_3 = pd.read_csv("dataset/02-16-2018.csv")
day_4 = pd.read_csv("dataset/02-20-2018.csv")
day_5 = pd.read_csv("dataset/02-21-2018.csv")
day_6 = pd.read_csv("dataset/02-22-2018.csv")
day_7 = pd.read_csv("dataset/02-23-2018.csv")
day_8 = pd.read_csv("dataset/02-28-2018.csv")
day_9 = pd.read_csv("dataset/03-01-2018.csv") 
day_10 = pd.read_csv("dataset/03-02-2018.csv")
```

```python
### 1- Concatenating the datasets
```

```python
dataset_1 = pd.concat([day_1, day_3 , day_5 , day_9, day_10], axis=0, ignore_index=True)
```

```python
dataset_1["Label"].unique()
```

```python
##2- Removing row with erroneous 'Label' label
```

```python
dataset_1.drop(dataset_1.loc[dataset_1["Label"] == "Label"].index, inplace=True)
```

```python
dataset_1["Label"].unique()
```

```python
print(dataset_1["Label"].value_counts()[['Benign']].sum())
print(dataset_1["Label"].value_counts()[['FTP-BruteForce']].sum())
print(dataset_1["Label"].value_counts()[['SSH-Bruteforce']].sum())
print(dataset_1["Label"].value_counts()[['DoS attacks-SlowHTTPTest']].sum())
print(dataset_1["Label"].value_counts()[['DoS attacks-Hulk']].sum())
print(dataset_1["Label"].value_counts()[['DDOS attack-LOIC-UDP']].sum())
print(dataset_1["Label"].value_counts()[['DDOS attack-HOIC']].sum())
print(dataset_1["Label"].value_counts()[['Infilteration']].sum())
print(dataset_1["Label"].value_counts()[['Bot']].sum())
```

```python
##3- Creating "Malicious" label to all other labels except 'Benign'
```

```python
dataset_1.replace(to_replace=[ 'FTP-BruteForce', 'SSH-Bruteforce',
       'DoS attacks-SlowHTTPTest', 'DoS attacks-Hulk',
       'DDOS attack-LOIC-UDP', 'DDOS attack-HOIC', 'Infilteration', 'Bot'
], value="Malicious", inplace=True)
```
